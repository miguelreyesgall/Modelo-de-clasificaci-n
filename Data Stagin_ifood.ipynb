# We will now review missing data

```python
df.notnull().apply(pd.Series.value_counts)
```

We notice that only the "Income" column contains "NaN" data, therefore, we can add to those cells data equal to the average of that column. 

```python
df['Income'] = df['Income'].fillna(df['Income'].median())
```

```python
# Confirm that there are no more NaN values
df.notnull().apply(pd.Series.value_counts)
```
